```python
import numpy as np
from matplotlib.gridspec import GridSpec
import matplotlib.pyplot as plt
from experiment.toolkits.figure_formatting.formatter import *
from experiment.toolkits.statistical_analysis.statistical_analysis import *

styles=[r'C:\Users\tqtraaqs\Desktop\tqtraaqs_git\matplotlibrc\styles\TQT-RAAQS.mplstyle']
```

```python
sigma = 8
yerr_all = []
yseries_all = []
xseries = list(range(0, 31, 2))
y_dist = []
for j in range(120, -20, -20):
    y = []
    for i in xseries:
        dist = abs(np.random.normal(j*(1 - np.exp(-i/3)), sigma, 1000))
        y.append(dist)
    y_dist.append(dist)
    
    y_mean, y_err = asymmetric_error(np.asarray(y), distribution='arbitrary')
    yseries_all.append(y_mean)
    yerr_all.append(y_err)

xseries_all = []
for i in range(6):
    xseries_all.append(xseries)
```

```python
fit_params={'equation':lambda x,a: a*(1 - np.exp(-x/3)), 
          'fit_xlim':[0,30,100]}

plot_params1={'xtitle': 'Scale',
             'xvariable': 'd',
             'xunits': '\u00b5m',
             'ytitle': 'Result',
             'yvariable': 'f(d)',
             'yticks': np.linspace(0, 125, 6, endpoint=True),
             'yunits': 'W/m$^2$',
             'caption_label': ['a', 'aps']
             }

plot_params2={
            'xtitle': 'Result',
            'xvariable': 'f(d)',
            'xunits': 'W/m$^2$',
            'ytitle': 'Frequency',
            'ytickdecimals': 2,
            'yticks': np.linspace(0, 0.06, 7, endpoint=True),
            'caption_label': ['b', 'aps']
            }

dist_series_params={
       'bins':30,
       'range': (0, 150)}
```
