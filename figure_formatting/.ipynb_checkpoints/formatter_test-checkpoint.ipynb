```python
import numpy as np
import matplotlib.pyplot as plt
from experiment.toolkits.figure_formatting.formatter import *
from experiment.toolkits.statistical_analysis.statistical_analysis import *
```

```python
sigma = 8
yerr_all = []
yseries_all = []
xseries = list(range(0, 31, 2))
y_dist = []
for j in range(120, -20, -20):
    y = []
    for i in xseries:
        dist = abs(np.random.normal(j*(1 - np.exp(-i/3)), sigma, 100))
        y.append(dist)
    y_dist.append(dist)
    
    y_mean, y_err = asymmetric_error(np.asarray(y), distribution='arbitrary')
    yseries_all.append(y_mean)
    yerr_all.append(y_err)

xseries_all = []
for i in range(6):
    xseries_all.append(xseries)
```

```python
fit_params={'equation':lambda x,a: a*(1 - np.exp(-x/3)), 
          'fit_xlim':[0,30,100]}

plot_params1={'xtitle': 'Scale',
             'xvariable': 'd',
             'xunits': '\u00b5m',
#              'xlabel_pad': -1,
             'ytitle': 'Result',
             'yvariable': 'f(d)',
             'yunits': 'W/m$^2$',
#              'ylabel_pad': -0.5,
             'caption_label': ['a', 'aps']
             }

plot_params2={
            'xtitle': 'Result',
            'xvariable': 'f(d)',
            'xunits': 'W/m$^2$',
#             'xlabel_pad': -1,
            'ytitle': 'Frequency',
            'ytickdecimals': 2,
            'xlim': (0, 150),
            'ylim': (0, 0.06),
            'caption_label': ['b', 'aps']
            }

dist_series_params={
       'bins':30,
       'range': (0, 150)}
```

```python
styles=['TQT-RAAQS']
phi = (1+np.sqrt(5))/2
width = 8.6/2.54
plot_width = 7.30/2.54
plot_xcoord = 1.15/2.54
v_plot_dist = 1.46/2.54
with plt.style.context(styles):
    # fig, ax = plt.subplots(2,2, figsize=(8.6, 8.6*2/(1 + np.sqrt(5))))
    fig, ax = plt.subplots(2,1)
    set_palette(ax[0], '2')
    for i in range(6):
        plot_scatter(ax[0], x=xseries_all[i], y=yseries_all[i], yerr=yerr_all[i], fit_params=fit_params)
    format_plot(ax[0], plot_params1)

    set_palette(ax[1], '2', plot_type='histogram')
    for i in range(6):
        plot_distribution(ax[1], x=y_dist[i], series_params=dist_series_params)
    format_plot(ax[1], plot_params2)
    format_axes(fig, ax)

    plt.savefig('test1.pdf')
    
# with plt.style.context(styles):
#     # fig, ax = plt.subplots(2,2, figsize=(8.6, 8.6*2/(1 + np.sqrt(5))))
#     fig, ax = plt.subplots()
#     set_palette(ax, '2')
#     for i in range(6):
#         plot_scatter(ax, x=xseries_all[i], y=yseries_all[i], yerr=yerr_all[i], fit_params=fit_params)
#     format_plot(ax, plot_params1)
    
#     height = width/phi
#     fig.set_figheight(height)
# #     fig1.set_size_inches(2,2)
#     ylim = ax.get_ylim()
#     xlim = ax.get_xlim()
#     aspect_ratio = (xlim[1]-xlim[0])/(ylim[1]-ylim[0])
#     ax.set_aspect(aspect_ratio/phi)
#     plt.subplots_adjust(left=plot_xcoord/width, bottom=v_plot_dist/height, right=1-(width-plot_xcoord-plot_width)/width, top=1, wspace=0, hspace=0)
#     plt.savefig('test2.pdf')

```

```python
import matplotlib.pyplot as plt

def adjustFigAspect(fig,aspect=1):
    '''
    Adjust the subplot parameters so that the figure has the correct
    aspect ratio.
    '''
    xsize,ysize = fig.get_size_inches()
    print(xsize, ysize)
    minsize = min(xsize,ysize)
    xlim = .4*minsize/xsize
    ylim = .4*minsize/ysize
    if aspect < 1:
        xlim *= aspect
    else:
        ylim /= aspect
    fig.subplots_adjust(left=.5-xlim,
                        right=.5+xlim,
                        bottom=.5-ylim,
                        top=.5+ylim)

fig = plt.figure(figsize=(8.6/2.54,20/2.54))
adjustFigAspect(fig,aspect=phi)
ax = fig.add_subplot(111)
ax.plot(range(10),range(10))
ax.set_ylabel('hello', fontsize=100)
plt.show()
```
